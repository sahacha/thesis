### Data Preprocess

```python
# There are some null value in content_list

for i in range(len(content_list)):
    for j in range(len(content_list[i])):
        print(content_list[i][j] == "")
```

```python
# Remove empty string in the content list

content_list = [list(filter(None, content_list[i])) for i in range(len(content_list))]
```

```python
# Use "แก้ไขข้อความเมื่อ" as a splitter and retain only the text before it

remove = "แก้ไขข้อความเมื่อ"
for i in range(len(content_list)):
    for j in range(len(content_list[i])):
        content_list[i][j] = re.sub(r'http\S+', '', content_list[i][j]) # remove website
        content_list[i][j] = content_list[i][j].split(remove, 1)[0] # split the sentence using "แก้ไขข้อความเมื่อ" and remove all the word after it
        content_list[i][j] = re.sub(r'\d+', '', content_list[i][j]) # remove number
        content_list[i][j] = content_list[i][j].translate(str.maketrans('', '', string.punctuation)) # remove punctuation
        content_list[i][j] = re.sub(r"\s+"," ",  content_list[i][j], flags = re.I) #remove multiple spaces within text
        content_list[i][j] = re.sub(r"\s+$", "", content_list[i][j]) # remove white spaces at the end of sentences
        content_list[i][j] = re.sub(r"^\s+", "", content_list[i][j])
```

```python
# Reove empty string for second time

content_list = [list(filter(None, content_list[i])) for i in range(len(content_list))]       
```

```python
# Create Dataframe

df_data = pd.DataFrame(subject)
df_data["comments"] = content_list
df_data = df_data.rename(columns={0: "subject"})
df_data.head()
```

## Thai Part-of-Speech Tagging

```python
# Import thai orchid

a={}
path = 'orchid97.crp.utf'
lines=codecs.open(path, 'r',encoding='utf8')
for line in lines:
  if len(re.findall('/', line)) == 1:
    lines1=line.split('/')
    a[lines1[0]]=re.sub('\n', "",lines1[1]) # เอา \n เมื่ออ่านไฟล์ออกก่อน
  else:
    lines1=0
    
with open('thaipos.pickle', 'wb+') as handle:
  pickle.dump(a, handle)

# Load orchid
with open('thaipos.pickle', 'rb') as f:
  thaipos = pickle.load(f)
```

```python
# Tokenize
from pythainlp import sent_tokenize, word_tokenize

# List comprehension
sent_token = [sent_tokenize(df_data["comments"][i][j]) 
for i in range(len(df_data["comments"])) 
for j in range(len(df_data["comments"][i]))]

word_token = [word_tokenize(df_data["comments"][i][j]) 
for i in range(len(df_data["comments"])) 
for j in range(len(df_data["comments"][i]))]
```

```python
#  Checking Spelling

from pythainlp.corpus import ttc  # Thai Textbook Corpus
from pythainlp.spell import NorvigSpellChecker

checker = NorvigSpellChecker(custom_dict=ttc.word_freqs())

# Visualize the current word and currented word
list_incorrect = {}
for i in range(len(word_token)):
    for j in range(len(word_token[i])):
        print("old:", word_token[i][j], "new:", checker.correct(word_token[i][j]))
        old = word_token[i][j]
        new = checker.correct(word_token[i][j])
        if old is not new:
            list_incorrect[i] = f"{old}; {new}"

```
